```python
import pandas as pd
import numpy as np
import os
```

```python
np.random.seed(42)
num_instances = 100
feature_size = [4, 4, 6, 8, 13, 22, 30, 34]
dataset_name = ['Iris',
                'Blood_Transfusion',
                'Vertebral-Column',
                'Pima',
                'Wine',
                'Parkinsons',
                'Breast_Cancer',
                'Ionosphere']
base_path = 'datasets/artificial/'
os.makedirs(base_path, exist_ok=True)

for i in range(len(feature_size)):
    data = np.random.rand(num_instances, feature_size[i])
    
    column_names = [f'f{j+1}' for j in range(feature_size[i])]
    
    df_exp = pd.DataFrame(data, columns=column_names)

    dataset_path = os.path.join(base_path, f'{dataset_name[i]}_artificial.csv')
    
    df_exp.to_csv(dataset_path, index=False)
    
    display(df_exp)
```
